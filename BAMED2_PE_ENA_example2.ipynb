#Create a manifest for the data

make sure we have changed working directory to /content/sample_data/

```
cd /content/sample_data/
```

Create a manifest file with two 3 columns (sample-id, forward-absolute-filepath,reverse-absolute-filepath)

```
!echo -e "sample-id\tforward-absolute-filepath\treverse-absolute-filepath" > manifest.tsv
```

```
ls *.gz
```

To remove underscores and hyphens (sed 's*_**g' is the part of the code that removes the _ and -)

```
%%bash
for file in *; do
    if [ -f "$file" ]; then
        mv "$file" `echo "$file" | sed 's*_**g'`
    fi
done
```

```
%%bash
for file in *; do
    if [ -f "$file" ]; then
        mv "$file" `echo "$file" | sed 's*-**g'`
    fi
done
```

The command below lists the fastq files, cuts the name after the fist R. We use this as sample-id. We create a file path for the forward and reverse reads, by printing the working directory, the sample-id and the file extension R1 or R2.fastq.gz for each of the files. The information is then added to the manifest.

The command below can also be used to create sample ids from file names with underscores or hyphens. 




```
%%bash
for f in `ls *fastq.gz | cut -d 'R' -f 1 | sort -u`; do n=`basename $f`; echo -e "${n}\t$PWD/$f"R1.fastq.gz"\t$PWD/$f"R2.fastq.gz" "; done >> manifest.tsv 
```

In the metadata we need to make a grouping column, in this case we will split sample_tile at the underscore (e-CD_01) to (e-CD) and (01)

```
metadata[['group', 'subject']]=metadata['sample_title'].str.split('_', expand=True)
```

```
metadata
```

The first column of the metadata needs to match the manifest

```
manifest=pd.read_table("manifest.tsv", index_col=[0])
```

```
manifest
```

```
id=manifest.index
id
```

```
sample_info=metadata['sample_title']
sample_info
```

We can see above that the order of the sample-id matches the order of sample_title. It is often easier to download the manifest and metadata from the ENA and compare/modify in excel.

To complete our metadata, we can give the metadata data frame the same index as the manifest, then save metadata as a text file.

```
metadata=metadata.set_index(id)
```

```
metadata
```

Finally, we can save our metadata file

```
metadata.to_csv('metadata.txt', header=True, index=True, sep='\t')
```

#Import data to Qiime2

```
!qiime tools import \
  --input-path manifest.tsv \
  --type 'SampleData[PairedEndSequencesWithQuality]' \
  --input-format PairedEndFastqManifestPhred33V2 \
  --output-path pe-demux.qza
```

```
!qiime demux summarize \
    --i-data  pe-demux.qza \
    --o-visualization pe-demux.qzv
```

The sequencing quality looks good, we will just truncate the end.

```
!qiime dada2 denoise-paired \
  --i-demultiplexed-seqs pe-demux.qza \
  --p-trunc-len-f 190 \
  --p-trunc-len-r 190 \
  --o-representative-sequences rep-seqs-dada2.qza \
  --o-table table-dada2.qza \
  --o-denoising-stats stats-dada2.qza
```

#Taxonomy

Make sure you have the correct version of the classifier

```
!wget \
  -O "gg-13-8-99-nb-classifier.qza" \
  "https://data.qiime2.org/2021.11/common/gg-13-8-99-nb-classifier.qza"
```

```
!qiime feature-classifier classify-sklearn \
  --i-reads rep-seqs-dada2.qza \
  --i-classifier gg-13-8-99-nb-classifier.qza \
  --o-classification taxonomy.qza
```

```
!qiime metadata tabulate \
  --m-input-file taxonomy.qza \
  --o-visualization taxonomy.qzv
```

#Phylogenetic tree

```
!qiime phylogeny align-to-tree-mafft-fasttree \
  --i-sequences rep-seqs-dada2.qza  \
  --o-alignment aligned-rep-seqs.qza \
  --o-masked-alignment masked-aligned-rep-seqs.qza \
  --o-tree unrooted-tree.qza \
  --o-rooted-tree rooted-tree.qza
```

The max depth was selected from table-dada2.qzv

```
!qiime diversity alpha-rarefaction \
  --i-table table-dada2.qza \
  --i-phylogeny rooted-tree.qza \
  --p-max-depth 12801 \
  --m-metadata-file metadata.txt \
  --o-visualization alpha-rarefaction.qzv
```

Here we perform the diversity analysis

```
#Core metrics
!qiime diversity core-metrics-phylogenetic \
  --i-phylogeny rooted-tree.qza \
  --i-table table-dada2.qza \
  --p-sampling-depth 12801 \
  --m-metadata-file metadata.txt \
  --output-dir core-metrics-results
```

```
!qiime diversity alpha-group-significance \
  --i-alpha-diversity core-metrics-results/faith_pd_vector.qza \
  --m-metadata-file metadata.txt \
  --o-visualization core-metrics-results/faith-pd-group-significance.qzv

!qiime diversity alpha-group-significance \
  --i-alpha-diversity core-metrics-results/evenness_vector.qza \
  --m-metadata-file metadata.txt \
  --o-visualization core-metrics-results/evenness-group-significance.qzv
```

```
!qiime diversity beta-group-significance \
  --i-distance-matrix core-metrics-results/weighted_unifrac_distance_matrix.qza \
  --m-metadata-file metadata.txt \
  --m-metadata-column group  \
  --o-visualization core-metrics-results/weighted-unifrac-significance.qzv \
  --p-pairwise
```
